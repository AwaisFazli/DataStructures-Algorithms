```python
expression = "2-3+4-5*6"
secStack = []
stack = []
postfix = []

expression = list(expression)

def isOperator(value):
    if value == "+" or value == "-" or value == "/" or value == "*" or value == "%" or value == "^":
        return True
    else:
        return False

def precedence(value):
    if value == "+" or value == "-":
        return 1
    elif value == "*" or value == "/":
        return 2
    elif value == "%":
        return 3
    elif value == "^":
        return 4

def isEmpty():
    if len(stack) == 0:
        return True
    else:
        return False

def top():
    return stack[-1]


def evaluate(operator):
    if operator == "*":
        value1 = secStack.pop()
        value2 = secStack.pop()
        secStack.append(int(value1) * int(value2))
    
    elif operator == "/":
        value1 = secStack.pop()
        value2 = secStack.pop()
        secStack.append(int(value1) / int(value2))
        
    elif operator == "-":
        value1 = secStack.pop()
        value2 = secStack.pop()
        secStack.append(int(value1) - int(value2))
        
    elif operator == "+":
        value1 = secStack.pop()
        value2 = secStack.pop()
        secStack.append(int(value1) + int(value2))

end = 0
i = 0
while i < len(expression):
    
    if end == 0:

        if not isOperator(expression[i]):
            postfix.append(expression[i])

        elif isOperator(expression[i]) and isEmpty():
            stack.append(expression[i])

        elif isOperator(expression[i]) and not isEmpty():
            if precedence(expression[i]) > precedence(top()):
                stack.append(expression[i])
            else:
                postfix.append(stack.pop())
                i -= 1

    if i == (len(expression)-1) and not isEmpty():
        postfix.append(stack.pop())
        i -= 1
        end += 1

        
    i += 1

    
print("Postfix Expression:",postfix)

i = 0
while i < len(postfix):
    if not isOperator(postfix[i]):
        secStack.append(postfix[i])
    
    
    elif isOperator(postfix[i]):
        evaluate(postfix[i])
    i += 1
        
print("Evaluated Result:",secStack[0])
```
